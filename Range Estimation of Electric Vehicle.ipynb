## Using Multivariate Linear Regression

```python
from sklearn import linear_model
```

```python
edata_train = edata[0:33:2]
edata_test = edata[0:33:5]

edata_train

```

```python
result = linear_model.LinearRegression()
result_fit =result.fit(edata_train[['Capacity of the Battery in Ampere Hour','Remaining Battery Percentage','SPEED']],edata_train['Driving Range in KM'])
```

```python
result.coef_  # forecast = coef1* capacity + coef2*remaining battery + coef3*SPEED +intercept
```

```python
result.intercept_
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
edata_test['Forecast'] = result.predict(edata_test[['Capacity of the Battery in Ampere Hour','Remaining Battery Percentage','SPEED']])
```

```python
edata_test['Actual'] = edata_test['Driving Range in KM'] 
```

```python
edata_test
```

 ## ACCURACY USING Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score

error = np.sqrt(mean_squared_error(edata_test['Actual'],edata_test['Forecast']))
```

```python
Percentage_Accuracy = 100 - error
Percentage_Accuracy
```
